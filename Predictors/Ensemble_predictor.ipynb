```python
import numpy as np
import pandas as pd
import os 
```

```python
# Aggregate Result Function 
def aggregate_result(row_series, condition=None):
    freq = row_series.value_counts().to_dict()
    
    _nonAMP = 0
    _AMP = 0
    _final_return = 0
    cutoff_threshold = 3 # RMSpop, Adam , ProtCNN, AmPEP30
    #print(freq)
    if 0 in  freq.keys():
        _AMP = freq[0]

    if 1 in freq.keys():
        _nonAMP = freq[1]  

    # print("AMP " + str(_AMP))
    # print("non AMP " + str(_nonAMP))
    # condition by pass rule if two models say  nonAMP
    # if(condition == "<=30"):
    if _nonAMP >= cutoff_threshold:
        #print("Rule fire")
        return 1 # NonAMP
    
    if _AMP > _nonAMP: # AMP
        _final_return = 0
    elif _AMP < _nonAMP: # NonAMP
        _final_return = 1
    else: #eql 
        pass
    
    return _final_return
```

# DECockroach

```python
# load result from each model 
path = "/mnt/vdb/DECockroach/cd100"
sample = "DECockroach.len15.MLResult.plk"
# RF
RF_df  = pd.read_pickle(path + "/result_2_RF/" + sample)
# XGBoost
XG_df  = pd.read_pickle(path + "/result_2_XG/" + sample)
# SVM
#SVM_df  = pd.read_pickle(path + "/result_2_SVM/" + sample)
# Custom CNN - RMSpop
cnnRMS_df  = pd.read_pickle(path + "/result_2_CNN_RMSpop/" + sample)
# Custom CNN - Adam
cnnAdam_df  = pd.read_pickle(path + "/result_2_CNN_Adam/" + sample)
# FastAI
FastAI_df  = pd.read_pickle(path + "/result_2_FastAI/" + sample)
# DNN
DNN_df  = pd.read_pickle(path + "/result_2_DNN/" + sample)
# ProtCNN
ProtCNN_df  = pd.read_pickle(path + "/result_2_ProtCNN/" + sample)
# AMpep30
ShortCNN_df  = pd.read_pickle(path + "/result_2_ShortCNN/" + sample)
```

```python
_tmp_df = pd.merge(RF_df,XG_df[['ID', 'class']],on='ID')
_tmp_df = _tmp_df.rename({'class_x': 'RF', 'class_y': 'XG'}, axis=1) 
_tmp_df = pd.merge(_tmp_df,cnnRMS_df[['ID', 'class']],on='ID')
_tmp_df = _tmp_df.rename({'class': 'RMSpop'}, axis=1) 
_tmp_df = pd.merge(_tmp_df,cnnAdam_df[['ID', 'class']],on='ID')
_tmp_df = _tmp_df.rename({'class': 'Adam'}, axis=1) 
_tmp_df = pd.merge(_tmp_df,FastAI_df[['ID', 'class']],on='ID')
_tmp_df = _tmp_df.rename({'class': 'FastAI'}, axis=1) 
_tmp_df = pd.merge(_tmp_df,DNN_df[['ID', 'class']],on='ID')
_tmp_df = _tmp_df.rename({'class': 'DNN'}, axis=1) 
_tmp_df = pd.merge(_tmp_df,ProtCNN_df[['ID', 'class']],on='ID')
_tmp_df = _tmp_df.rename({'class': 'ProtCNN'}, axis=1) 
_tmp_df = pd.merge(_tmp_df,ShortCNN_df[['ID', 'class']],on='ID')
_tmp_df = _tmp_df.rename({'class': 'ShortCNN'}, axis=1) 
_tmp_df.sort_values(by=['ID'], inplace=True, ascending=True)
_tmp_df
```

```python
final_df = pd.DataFrame()
final_df["ID"] =_tmp_df["ID"]
final_df["length"] =_tmp_df["length"]
_tmp_df.drop(columns=['ID','length'],inplace =True)
final_df["class"] = _tmp_df.apply(lambda row: aggregate_result(row) , axis = 1)
final_df
```

```python
path = "/mnt/vdb/DECockroach/cd100/result_2_Ensemble"
if not os.path.exists(path):
    os.makedirs(path)
final_df.to_pickle(path+"/DECockroach.len15.MLResult.plk")
```
