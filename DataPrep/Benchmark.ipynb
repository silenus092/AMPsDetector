```python
import pandas as pd
import numpy as np
from Bio import SeqIO
```

```python
# construct badlist
amp_fasta = "../datasets/thesis/trainingset2/benchmark_set/920_920/AMPs.fasta"
non_amp_fasta = "../datasets//thesis/trainingset2/benchmark_set/920_920/NonAMPs.fasta"
```

```python
with open(amp_fasta) as fasta_file:  # Will close handle cleanly
    identifiers = []
    lengths = []
    sequence = []
    for seq_record in SeqIO.parse(fasta_file, 'fasta'):  # (generator)
        identifiers.append(seq_record.id)
        sequence.append(str(seq_record.seq))
        lengths.append(len(seq_record.seq))

AMPs_df = pd.DataFrame(list(zip(identifiers, sequence, lengths)), columns =['ID', 'Sequence', 'length']) 
AMPs_df
```

```python
AMPs_df.length.describe()
```

```python
with open(non_amp_fasta) as fasta_file:  # Will close handle cleanly
    identifiers = []
    lengths = []
    sequence = []
    for seq_record in SeqIO.parse(fasta_file, 'fasta'):  # (generator)
        identifiers.append(seq_record.id)
        sequence.append(str(seq_record.seq))
        lengths.append(len(seq_record.seq))

NonAMPs_df = pd.DataFrame(list(zip(identifiers, sequence, lengths)), columns =['ID', 'Sequence', 'length']) 
NonAMPs_df
```

```python
NonAMPs_df.length.describe()
```

```python
AMPs_df["class"] = 0
NonAMPs_df["class"] = 1
```

```python
ready_df = pd.concat([AMPs_df,NonAMPs_df], axis=0,ignore_index=True)
ready_df
```

```python
ready_df.drop_duplicates(subset=['Sequence'],inplace=True)
ready_df
```

```python
print(len(ready_df[ready_df['class'] == 0]))
print(len(ready_df[ready_df['class'] == 1]))
```
